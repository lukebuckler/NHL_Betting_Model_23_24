```python
import pandas as pd
import os

# Path to the directory containing your daily game files
directory_path = 'C:\\Users\\luken\\Desktop\\NHL_Model\\23_24_results'

# List all CSV files in the directory
csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]

# Initialize an empty list to hold the dataframes
daily_games_dfs = []

# Iterate over each csv file in the directory
for file_name in csv_files:
    # Construct the full file path
    file_path = os.path.join(directory_path, file_name)
    
    # Load the CSV file into a DataFrame
    daily_df = pd.read_csv(file_path)
    
    # Add the DataFrame to our list
    daily_games_dfs.append(daily_df)

# Concatenate all the DataFrames in the list into one DataFrame
all_daily_games_df = pd.concat(daily_games_dfs, ignore_index=True)

# Now all_daily_games_df contains all the data from your daily game files

```

```python
season_results = pd.read_csv('23_24_Games/23_24_games.csv')
season_results.head()
```

```python
season_results['GameInfo'] = season_results['Game'].str.split(' - ').str[0] + ' - ' + \
                                season_results['Game'].str.split(' - ').str[1].str.split(' ').str[0] + ', ' + \
                                season_results['Game'].str.split(' - ').str[1].str.split(' ').str[2]

# Display the first few rows to verify the new column
print(season_results[['Game', 'GameInfo']].head())
```

```python
games_info_df = season_results[['Game', 'GameInfo']].copy()

# Display the first few rows of the new DataFrame to verify
print(games_info_df.head())
```

```python
games_info_df[['Date', 'AwayTeam', 'AwayScore', 'HomeTeam', 'HomeScore']] = games_info_df['Game'].str.extract(
    r'(\d{4}-\d{2}-\d{2}) - (\w+) (\d+), (\w+) (\d+)'
)

# Convert scores to numeric values to facilitate comparison
games_info_df['AwayScore'] = pd.to_numeric(games_info_df['AwayScore'])
games_info_df['HomeScore'] = pd.to_numeric(games_info_df['HomeScore'])

# Initialize the 'HomeResult' column to 'Draw'
games_info_df['HomeResult'] = 'Draw'

# Set 'HomeResult' to 'Won' where the home team scored more than the away team
games_info_df.loc[games_info_df['HomeScore'] > games_info_df['AwayScore'], 'HomeResult'] = 'Won'

# Set 'HomeResult' to 'Lost' where the home team scored less than the away team
games_info_df.loc[games_info_df['HomeScore'] < games_info_df['AwayScore'], 'HomeResult'] = 'Lost'

# Display the first few rows to verify the new columns and 'HomeResult' values
print(games_info_df.head())
```

```python
games_info_df = games_info_df.drop_duplicates(subset=['GameInfo'])
print(games_info_df.head())
```

```python
combined_df = pd.merge(games_info_df, all_daily_games_df, on='GameInfo', how='inner')

# Display the first few rows of the combined DataFrame to verify
print(combined_df.head())
```

```python
# Specify the path and name of the output CSV file
output_file_path = 'combined_games_info.csv'

# Save the combined DataFrame to a CSV file, without including the index
combined_df.to_csv(output_file_path, index=False)

# If you're running this in a Jupyter notebook or similar, you might not see any output after running this command.
# The absence of errors means the file has been successfully written.
```

```python
output_file_path = 'betting_anal.csv'

# Save the combined DataFrame to a CSV file, without including the index
bet_anal.to_csv(output_file_path, index=False)
```

```python
import numpy as np

# Assuming 'df' is your DataFrame, and you're comparing 'ColumnA' and 'ColumnB'
# Create a new column 'NewColumn' that has a 1 if 'ColumnA' > 'ColumnB', and NaN if not
bet_anal['AwayBet'] = np.where(bet_anal['Away Probability'] > bet_anal['Away Team Implied Probability'], 1, 0)
bet_anal['HomeBet'] = np.where(bet_anal['Home Probability'] > bet_anal['Home Team Implied Probability'], 1, 0)
# Display the first few rows to verify the new column
print(bet_anal.head())
```

```python
bet_anal['AwayBetWon'] = np.where(bet_anal['HomeResult'] == "Lost", 1, 
                            np.where(bet_anal['HomeResult'] == "Won", 0, np.nan))
```

```python
bet_anal['HomeBetWon'] = np.where(bet_anal['HomeResult'] == "Won", 1, 
                            np.where(bet_anal['HomeResult'] == "Lost", 0, np.nan))
bet_anal.head()
```

```python
# Assuming the DataFrame is named bet_anal
bet_anal['total_results'] = (bet_anal['AwayBet'] * bet_anal['AwayBetWon']) + (bet_anal['HomeBet'] * bet_anal['HomeBetWon'])
bet_anal['total_bets'] = (bet_anal['AwayBet']) + (bet_anal['HomeBet'])
# Display the first few rows to verify the new column
print(bet_anal[['AwayBet', 'AwayBetWon', 'HomeBet', 'HomeBetWon', 'total_results', 'total_bets']].head())
```

```python
bet_anal['Away Team Implied Probability'] = bet_anal['Away Team Implied Probability'].astype(float)
bet_anal['Home Team Implied Probability'] = bet_anal['Home Team Implied Probability'].astype(float)
bet_anal['AwayBet'] = pd.to_numeric(bet_anal['AwayBet'], errors='coerce')
bet_anal['HomeBet'] = pd.to_numeric(bet_anal['HomeBet'], errors='coerce')
```

```python
columns_to_convert = ['AwayBet', 'HomeBet', 'AwayBetWon', 'HomeBetWon']

# Convert each specified column to float using pd.to_numeric()
for column in columns_to_convert:
    bet_anal[column] = pd.to_numeric(bet_anal[column], errors='coerce')

# Display the data types of the columns to verify the conversion
print(bet_anal[columns_to_convert].dtypes)
```

```python
bet_anal['total_return'] = (bet_anal['AwayBet']*(1/('Away Team Implied Probability')) * bet_anal['AwayBetWon']) + (bet_anal['HomeBet']*(1/('Home Team Implied Probability')) * bet_anal['HomeBetWon'])
```

```python
bet_anal['Away Team Implied Probability'] = pd.to_numeric(bet_anal['Away Team Implied Probability'], errors='coerce')
bet_anal['Home Team Implied Probability'] = pd.to_numeric(bet_anal['Home Team Implied Probability'], errors='coerce')

# Calculate 'total_return' with corrected column references and operations
bet_anal['total_return'] = (bet_anal['AwayBet'] * (1 / bet_anal['Away Team Implied Probability']) * bet_anal['AwayBetWon']) + \
                           (bet_anal['HomeBet'] * (1 / bet_anal['Home Team Implied Probability']) * bet_anal['HomeBetWon'])
```

```python
total_results_sum = bet_anal['total_results'].sum()
total_bets_sum = bet_anal['total_bets'].sum()
total_return_sum = bet_anal['total_return'].sum()
```
